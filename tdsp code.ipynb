**Step 2:** Let's begin by creating a chart that displays the average number of crashes per hour of the day. This will help us understand whether additional factors are contributing to crashes - i.e. rush hour, school dismissal time, night fall, etc.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = "/content/drive/MyDrive/Motor_Vehicle_Collisions_-_Crashes_20250106.csv"
data = pd.read_csv(file_path)

# Convert 'CRASH DATE' and 'CRASH TIME' to datetime
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])
data['CRASH TIME'] = pd.to_datetime(data['CRASH TIME'], format='%H:%M')

# Time of Day Analysis
data['Hour of Day'] = data['CRASH TIME'].dt.hour

# Group by 'Hour of Day' and calculate the average number of crashes per hour
average_crashes_per_hour = data.groupby('Hour of Day').size() / data['Hour of Day'].nunique()

# Plot the average number of crashes
plt.figure(figsize=(12, 6))
sns.barplot(x=average_crashes_per_hour.index, y=average_crashes_per_hour)
plt.title('Average Number of crashes per Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Number of crashes')
plt.xticks(range(0, 24))
plt.show()


````

**Step 4**: Apply time series decomposition to review trends, seasonality, and residuals. New to time series analysis? Review the [Time Series Flashcard video series](https://youtube.com/playlist?list=PLNs9ZO9jGtUAqd0CNKySksPJJaOqYPMvk&feature=shared) here to learn about trends, components, and further analysis!

````python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Count the number of crashes per day, group by CRASH DATE
daily_crashes = data.groupby(data['CRASH DATE']).size()

# Set plot style
sns.set(style="darkgrid")

# Plot the daily crashes time series
plt.figure(figsize=(15, 6))
plt.plot(daily_crashes, label='Daily crashes')
plt.title('Daily Motor Vehicle Collisions in NYC')
plt.xlabel('Day')
plt.ylabel('Crashes')
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(daily_crashes, model='additive', period=365)

# Plot the decomposed components
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
decomposition.seasonal.plot(ax=ax2)
ax2.set_title('Seasonality')
decomposition.resid.plot(ax=ax3)
ax3.set_title('Residuals')
plt.tight_layout()
plt.show()
````

The visualizations above provide valuable insights into the time series of daily motor vehicle collisions in New York City:

1. Time Series Plot: This shows the number of daily crashes over time. You might observe long-term trends, seasonal patterns, or significant outliers.

2. Decomposed Components:
  
    2.1 Trend: This graph shows the long-term trend in the data, which can indicate whether crashes are increasing, decreasing, or stable over time.

    2.2 Seasonality: This reveals any regular patterns that repeat over a specific period, such as yearly. It helps identify times of the year with higher or lower crash frequencies.

    2.3 Residuals: These are the irregular components that cannot be attributed to the trend or seasonality. They might include random or unpredictable fluctuations.

**Step 3:** Next, we will begin "Severity Mapping." We'll plot crashes on the map and code them based on severity, distinguishing between crashes that resulted in injuries and those that led to fatalities. This will give us a visual representation of where severe crashes tend to occur. </br>

You may initially want to code these incidents by using different colors, but it's important to think about your map and how accessible it is. Will a color-coded map be easy to read for everyone? Let's make a map that's inclusive for people with color blindness by also creating differently-shaped markers (squares, circles, and triangles) for crashes, injuries, and fatalities.

````python
#TODO: Continue building your heatmap
# Sample a subset of the data for visualization
sample_data_severity = data_geo.sample(n=1000, random_state=75)

# Create a base map
m_severity = folium.Map(location=[40.730610, -73.935242], zoom_start=10)

# Add crashes to the map with color coding and shape coding based on severity
for index, row in sample_data_severity.iterrows():
    if row['NUMBER OF PERSONS KILLED'] > 0:
        color = "RED"  # Fatalities

        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=3,
          radius=5,
          gradient = False,
          color=color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)


    elif row['NUMBER OF PERSONS INJURED'] > 0:
        color = "BLUE"  # Injuries
        folium.CircleMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          radius=5,
          color=color,
          fill=True,
          fill_color=color
       ).add_to(m_severity)
    else:
        color = "GRAY"  # No injuries or fatalities
        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=4,
          radius=5,
          gradient = False,
          color=color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)


m_severity.save("severity.html")
````

## **Research Question 1: How do weather conditions influence the number of motorist and pedestrian injuries in crashes?**

````python
# Load the crash table
crash_df = pd.read_csv("/content/drive/MyDrive/Motor_Vehicle_Collisions_-_Crashes_20250106.csv")

# Load the weather dataset
weather_df = pd.read_csv("/content/drive/MyDrive/NYC_Weather_2016_2022_data.csv")

# Convert 'time' column in weather data to datetime format if not done already
weather_df['time'] = pd.to_datetime(weather_df['time'])

# Extract date and time from the weather data
weather_df['CRASH DATE'] = weather_df['time'].dt.strftime('%m/%d/%Y')  # Format as MM/DD/YYYY
weather_df['CRASH TIME'] = weather_df['time'].dt.strftime('%H:%M')

weather_df.head()
````

````python
# Merge crash and weather data on CRASH_DATE and CRASH_TIME
merged_df = pd.merge(crash_df, weather_df, how='inner',
                     left_on=['CRASH DATE', 'CRASH TIME'],
                     right_on=['CRASH DATE', 'CRASH TIME'])
merged_df.info()
merged_df.head()
````

````python
# Create a scatter plot for number of injuries vs. weather conditions

# Scatter plot for Motorist Injuries vs Temperature
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x='temperature_2m (°C)', y='NUMBER OF MOTORIST INJURED',
                hue='temperature_2m (°C)', palette='cubehelix', s=100)
plt.title("Motorist Injuries vs Temperature", fontsize=16)
plt.xlabel("Temperature (°C)", fontsize=14)
plt.ylabel("Number of Motorist Injured", fontsize=14)
plt.show()

# Scatter plot for Pedestrian Injuries vs Precipitation
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x='precipitation (mm)', y='NUMBER OF PEDESTRIANS INJURED',
                hue='precipitation (mm)', palette='cubehelix', s=100)
plt.title("Pedestrian Injuries vs Precipitation", fontsize=16)
plt.xlabel("Precipitation (mm)", fontsize=14)
plt.ylabel("Number of Pedestrians Injured", fontsize=14)
plt.show()

# Scatter plot for Motorist Injuries vs Rain
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x='rain (mm)', y='NUMBER OF MOTORIST INJURED',
                hue='rain (mm)', palette='cubehelix', s=100)
plt.title("Motorist Injuries vs Rain", fontsize=16)
plt.xlabel("Rain (mm)", fontsize=14)
plt.ylabel("Number of Motorist Injured", fontsize=14)
plt.show()

# Scatter plot for Pedestrian Injuries vs Cloudcover
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x='cloudcover (%)', y='NUMBER OF PEDESTRIANS INJURED',
                hue='cloudcover (%)', palette='cubehelix', s=100)
plt.title("Pedestrian Injuries vs Cloudcover", fontsize=16)
plt.xlabel("Cloudcover (%)", fontsize=14)
plt.ylabel("Number of Pedestrians Injured", fontsize=14)
plt.show()

# Scatter plot for Motorist Injuries vs Windspeed
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x='windspeed_10m (km/h)', y='NUMBER OF MOTORIST INJURED',
                hue='windspeed_10m (km/h)', palette='cubehelix', s=100)
plt.title("Motorist Injuries vs Windspeed", fontsize=16)
plt.xlabel("Windspeed (km/h)", fontsize=14)
plt.ylabel("Number of Motorist Injured", fontsize=14)
plt.show()

# Scatter plot for Pedestrian Injuries vs Winddirection
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x='winddirection_10m (°)', y='NUMBER OF PEDESTRIANS INJURED',
                hue='winddirection_10m (°)', palette='cubehelix', s=100)
plt.title("Pedestrian Injuries vs Winddirection", fontsize=16)
plt.xlabel("Winddirection (°)", fontsize=14)
plt.ylabel("Number of Pedestrians Injured", fontsize=14)
plt.show()
````

## **Research Question 3: What types of vehicle conditions (e.g., vehicle damage, pre-crash behaviors), and the presence of safety equipment (e.g., seatbelts, helmets) are most likely to be involved in crashes in different boroughs?**

````python
# Load the vehicle dataset
vehicle_df = pd.read_csv("/content/drive/MyDrive/Motor_Vehicle_Collisions_-_Vehicles_20250117.csv")


# Merge crash and weather data on CRASH_DATE and CRASH_TIME
merged1_df = pd.merge(crash_df, vehicle_df, on='COLLISION_ID', how='inner')

merged1_df.head()
````

````python
# Load the person table
person_df = pd.read_csv("/content/drive/MyDrive/Motor_Vehicle_Collisions_-_Person_20250117.csv")

# Merge crash and person data on COLLISION_ID
merged2_df = pd.merge(crash_df, person_df, on='COLLISION_ID', how='inner')

merged2_df.head()
````

````python
# 1. Grouped Bar Plot: Vehicle Damage by Borough
plt.figure(figsize=(10, 5))
vehicle_damage_borough = pd.crosstab(merged1_df['VEHICLE_DAMAGE'], merged1_df['BOROUGH'])

# Converting to a proportion for better comparison
vehicle_damage_borough = vehicle_damage_borough.div(vehicle_damage_borough.sum(axis=0), axis=1)

vehicle_damage_borough.plot(kind='bar', figsize=(10, 5), stacked=False, colormap='cubehelix')

plt.title("Distribution of Vehicle Damage by Borough", fontsize=16)
plt.xlabel("Vehicle Damage Level", fontsize=14)
plt.ylabel("Proportion of Crashes", fontsize=14)
plt.legend(title="Borough")
plt.show()

# 2. Line Plot: Pre-Crash Behaviors by Borough
plt.figure(figsize=(10, 5))
pre_crash_borough = pd.crosstab(merged1_df['PRE_CRASH'], merged1_df['BOROUGH'])

# Normalize to proportions
pre_crash_borough = pre_crash_borough.div(pre_crash_borough.sum(axis=0), axis=1)

# Plot each borough as a line
pre_crash_borough.T.plot(kind='line', figsize=(10, 5), marker='o')

plt.title("Pre-Crash Behaviors by Borough", fontsize=16)
plt.xlabel("Borough", fontsize=14)
plt.ylabel("Proportion of Crashes", fontsize=14)
plt.legend(title="Pre-Crash Behavior", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 3. Heatmap: Safety Equipment Usage by Borough
plt.figure(figsize=(10, 5))
safety_equipment_borough = pd.crosstab(merged2_df['SAFETY_EQUIPMENT'], merged2_df['BOROUGH'], normalize='columns')

sns.heatmap(safety_equipment_borough, annot=True, cmap='cubehelix', fmt=".2f")

plt.title("Safety Equipment Usage by Borough", fontsize=16)
plt.xlabel("Borough", fontsize=14)
plt.ylabel("Safety Equipment", fontsize=14)
plt.show()
````
